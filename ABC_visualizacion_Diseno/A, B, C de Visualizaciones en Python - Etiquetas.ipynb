```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tiendas_procesado.csv', index_col='fecha')
df
```

```python
import matplotlib.pyplot as plt
```

```python

plt.style.use(
    'ggplot' 
)

ax = df.plot.line(
    figsize=(10,5),
    title="Ventas de 3 tipos de productos en el ao 2015",
)

ax.set_xlabel(
    'Perodo',
    loc='center', #alineamiento
)

ax.set_ylabel(
    'Ventas (en millones de dolares)',
    loc='center', #alineamiento, top, bottom
)

```

```python
import seaborn as sns
```

```python
#cambiar el tamao del canvas
plt.figure(
    figsize=(8,5)
)

plt.title('Ventas!!!')
plt.xlabel("Periodo")
plt.ylabel("Ventas (en millones de dolares)")


sns.set_style("ticks") 
sns.set_theme("paper")


ax = sns.lineplot(
    data=df
)

ax.set(
    xlabel="X LABEL",
    ylabel="Y LABEL",
    title="TITULO"
)


#ax.set_xlabel('Periodo')

```

```python
import plotly.express as px
```

https://plotly.com/python/templates/

```python
fig = px.line(
    df,
    width=800,# medida en px
    height=500,
    template='ggplot2', # styles por temas
    title='Ventas de 3 tipos de productos por ao',
    labels={ # funciona con los title que ya tiene como llave
        'fecha':'Periodo',
        'value':'Ventas por ao en millones de $',
        'variable':'Tipo de producto'
    }
)

# otra forma mas intuitiva
fig.update_layout(
    xaxis_title="PERIODO!!!",
    yaxis_title="VALORES!!",
)

fig.show()
```

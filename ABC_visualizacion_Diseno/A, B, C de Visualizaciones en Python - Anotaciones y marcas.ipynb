# Estilos de Etiquetas y texto


```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tiendas_procesado.csv', index_col='fecha')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

plt.style.use(
    'ggplot' 
)

ax = df.plot.line(
    figsize=(10,5)
)

# https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text
ax.set_title(
     "Ventas de 3 tipos de productos en el año 2015",
    fontdict={ # diccionario para cambiar los estilos
        'size':20,
        'weight':'bold',
        
    }
)

ax.set_yticks(
    [500000,1000000]
)

ax.set_yticklabels(
    ['500000', '1000000'],
    fontdict={
        'size':10
    }
)

ax.set_xticks( np.arange( 0,len(df) ) )

ax.set_xticklabels(
    df.index.values,
    fontdict={
        'name': 'Times new roman',
        #'family':'serif',
        'color':'blue',
        'weight':'bold',
        'size':10
        
    }
    
)



ax.set_xlabel(
    'Período'
)

ax.set_ylabel(
    'Ventas (en millones de dolares)',
    fontdict={
        'font':'comic sans MS',
        'color':'red',
        'size':22,
        
    }
)


# metodo glbal de matplotlib si no redefinimos los ticks del eje Y
plt.yticks(# recibe directamente las propiedades
    color='blue',
    name='Times new roman'
)
plt.xticks(
    color='green',
    name='Times new roman'
)
```

```python
import plotly.express as px
```

https://plotly.com/python/templates/

```python
fig = px.line(
    df,
    width=800,# medida en px
    height=500,
    template='ggplot2', # styles por temas
    title='Ventas de 3 tipos de productos por año',

)


fig.update_layout(
    xaxis_title="PERIODO!!!",
    yaxis_title="VALORES!!",
    #personalizacion de textos
    font_color="blue",
    font_family="Courier New",
    # cambiar color del texto de la leyenda
    legend_font_color='black',
    legend_title_font_color='green',
    title_font_color='red',
    title_font_family='Times New Roman'
)


fig.update_xaxes(
    tickangle=45,
    tickvals=df.index.values[::2],
    tickmode='array',
    #personalizacion textos
    title_font_family='Arial',
    title_font_color='black',
    title_font_size=10,
    # personalizacion ticks
    tickfont={
        'family':'Rockwell',
        'color':'brown',
        'size': 10
    }
)

fig.update_yaxes(
    
    dtick=250000,
 
)

fig.show()
```

# Canvas y Temas 

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tiendas_procesado.csv', index_col='fecha')
df
```

```python
import matplotlib.pyplot as plt
```

```python
# cambio global en matplot lib
plt.rcParams["figure.figsize"] = (15,3) #lo vuelve predeterminado
# afecta tambien a las otras librerias como seaborn
```

Referencia a temas
https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

```python
# añadir style por tema
plt.style.use(
    'bmh' # ggplot tambien es bueno
)

ax = df.plot.line( # retorna una referencia al objeto del grafico
    # Dimensiones
    figsize=(12,3) # ancho, alto, son pulgadas
)

# utilizamos la variable con la referencia
ax.xaxis.grid(
    color='blue',
    alpha=0.2,
    linewidth=0.5
    #visible=False # para que funcione hay que quitarle los otros parametros
)

ax.yaxis.grid(
    #visible=False
    color='#28bf50',
    linestyle='-.', #https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle



)
```

```python
import seaborn as sns
```

```python
#cambiar el tamaño del canvas
plt.figure(
    figsize=(8,5)
)
# styles especificos de seaborn
sns.set_style("ticks") #dark, darkgrid etc
sns.set_theme("paper") # talk hace que sea mas eficiente/grande a la hora de leerlo
# paper mas pequeño para articulo cientifico

# metodo global
plt.grid(
    visible=False,
    axis="x", # quitmaos grid x
)

plt.grid(
    axis="y", # quitmaos grid x
    color="red",
    alpha=0.2
)


ax = sns.lineplot( # cogemos la referencia con ax
    data=df
)


```

```python
import plotly.express as px
```

https://plotly.com/python/templates/

```python
fig = px.line(
    df,
    width=800,# medida en px
    height=500,
    template='plotly_dark', # styles por temas
)

# modificamos los ejes del grid
fig.update_xaxes(showgrid=False)
fig.update_yaxes(
    gridwidth=1,
    gridcolor='white', #nombre o HEX
    griddash='solid',
    minor_griddash='dot' # subgrid
)
```

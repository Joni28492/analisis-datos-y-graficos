# Leyendas


```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tiendas_procesado.csv', index_col='fecha')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

plt.style.use(
    'ggplot' 
)

ax = df.plot.line(
    figsize=(10,4)
)


ax.set_title(
     "Ventas de 3 tipos de productos en el año 2015",
    fontdict={
        'size':20,
        'weight':'bold',
        
    }
)


ax.set_xticks( np.arange( 0,len(df) ) )

ax.set_xticklabels(
    df.index.values,
    fontdict={
        'color':'black',
        'size':8
        
    }
    
)



ax.set_xlabel(
    'Período',
    fontdict={
        'color':'black',
        'size':12
    }
)

ax.set_ylabel(
    'Ventas (en millones de dolares)',
    fontdict={
        'color':'black',
        'size':12,
        
    }
)


ax.annotate(
    'Noviembre OJO!', 
    xy=(10,1250000),
    xytext=(10, 1000000),
    color='grey',
    horizontalalignment='center',
    arrowprops={
        'width':1,
        'color':'black',
        'headwidth': 5,
        'headlength':5 
    }
)


ax.legend(
    title="Titulo de producto",
    labels=["Lacteos", "Carne", "Comida de mar"],
    fontsize=10,
    # si se mueve es para tapar la menor parte del grafico
    loc='upper left',# punto de anclaje de la leyenda, que lo utilizamos para fijar las cordenadas
    bbox_to_anchor=(0.25,-0.15),# no son cordenadas son % de 0 a 1 
    # podemos poner valores superiores a 1, adminete negativos
    labelcolor=["red", 'blue', 'purple'],
    # es como el font dict
    # title_fontproperties={'size':17},
    # alignment='left', # alineamiento de las etiquetas
    ncol=3, # numero de columnas, como el dflex
    
)
```

```python
import seaborn as sns
```

```python
#cambiar el tamaño del canvas
plt.figure(
    figsize=(8,5)
)

plt.title('Ventas!!!')
plt.xlabel("Periodo")
plt.ylabel("Ventas (en millones de dolares)")


sns.set_style("ticks") 
sns.set_theme("paper")


ax = sns.lineplot(
    data=df
)


# linea horizontal que recorre todo el grafico
ax.axhline(
    1300000,
    linestyle='dashed',
    linewidth=0.7,
    alpha=0.7,
    color='red'
)

#ax.axvline(#linea vertical)
# anotacion para la linea
ax.annotate(
    'Umbral',
    xy=(11,1250000),
    horizontalalignment='center',
    color='red',
    fontstyle='italic',
    alpha=0.8,
    size=8
)

ax.legend().set_visible(False) # ocultar legenda


```

```python
import plotly.express as px
```

https://plotly.com/python/templates/

```python
fig = px.line(
    df,
    width=800,# medida en px
    height=500,
    template='ggplot2', # styles por temas
    title='Ventas de 3 tipos de productos por año',

)


fig.update_layout(
    xaxis_title="PERIODO!!!",
    yaxis_title="VALORES!!",
    #personalizacion de textos
    font_color="blue",
    font_family="Courier New",
    # cambiar color del texto de la leyenda
    legend_font_color='black',
    legend_title_font_color='green',
    title_font_color='red',
    title_font_family='Times New Roman'
)


fig.update_xaxes(
    tickangle=45,
    tickvals=df.index.values[::2],
    tickmode='array',
    #personalizacion textos
    title_font_family='Arial',
    title_font_color='black',
    title_font_size=10,
    # personalizacion ticks
    tickfont={
        'family':'Rockwell',
        'color':'brown',
        'size': 10
    }
)

fig.update_yaxes(
    
    dtick=250000,
 
)


fig.add_annotation( # viene ya con linea no con =>
    x='2015-11', 
    y=1300000,
    text="Noviembre!",
    opacity=0.5,
    font={
        'size':12,
        'color':'black'
    },
    showarrow=True,
    arrowsize=1,
    arrowwidth=2,
    arrowhead=2# tipo de flecha por numero de 0-8
)

# agregamos linea horizontal
fig.add_hline(
    y=1300000,
    annotation_text="Limite",
    annotation_position="top left",# respecto de la linea
    annotation_font_color='red',
    annotation_font_size=10,
    line_dash="dot",
    opacity=0.8,
    line_width=1,
    line_color='red'
)

# personalizar leyebda

fig.update_layout(
    legend_title_text='Leyenda',
    legend={# estilos de la leyenda
        # posicionamiento
        'y': 0.5,
        'x': 1.05,
        'font':{
            'family':'Courier',
            'size':12,
            'color':'Black'
        },
        'bgcolor': 'LightGrey',
        'bordercolor':'Black',
        'borderwidth': 2
    }

)


fig.show()
```

En plotly no tenemos la misma libertad para hacer subgraficos


https://plotly.com/python/templates/

```python
fig = px.line(
    df,
    width=800,# medida en px
    height=500,
    template='ggplot2', # styles por temas
    title='Ventas de 3 tipos de productos por ao',

)


fig.update_layout(
    xaxis_title="PERIODO!!!",
    yaxis_title="VALORES!!",
    #personalizacion de textos
    font_color="blue",
    font_family="Courier New",
    # cambiar color del texto de la leyenda
    legend_font_color='black',
    legend_title_font_color='green',
    title_font_color='red',
    title_font_family='Times New Roman'
)


fig.update_xaxes(
    tickangle=45,
    tickvals=df.index.values[::2],
    tickmode='array',
    #personalizacion textos
    title_font_family='Arial',
    title_font_color='black',
    title_font_size=10,
    # personalizacion ticks
    tickfont={
        'family':'Rockwell',
        'color':'brown',
        'size': 10
    }
)

fig.update_yaxes(
    
    dtick=250000,
 
)


fig.add_annotation( # viene ya con linea no con =>
    x='2015-11', 
    y=1300000,
    text="Noviembre!",
    opacity=0.5,
    font={
        'size':12,
        'color':'black'
    },
    showarrow=True,
    arrowsize=1,
    arrowwidth=2,
    arrowhead=2# tipo de flecha por numero de 0-8
)

# agregamos linea horizontal
fig.add_hline(
    y=1300000,
    annotation_text="Limite",
    annotation_position="top left",# respecto de la linea
    annotation_font_color='red',
    annotation_font_size=10,
    line_dash="dot",
    opacity=0.8,
    line_width=1,
    line_color='red'
)

# personalizar leyebda

fig.update_layout(
    legend_title_text='Leyenda',
    legend={# estilos de la leyenda
        # posicionamiento
        'y': 0.5,
        'x': 1.05,
        'font':{
            'family':'Courier',
            'size':12,
            'color':'Black'
        },
        'bgcolor': 'LightGrey',
        'bordercolor':'Black',
        'borderwidth': 2
    }

)


fig.show()
```

para hacer subgraficos con plotly tenemos otro paquete

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
fig = make_subplots(rows=3, cols=1 )

fig.append_trace(
    go.Scatter(# aadimos grafico de tipo scatter, el de lineas
        x=df.index.values,
        y=df['DAIRY'],
        name='LACTEOS'
    ),row=1 , col=1
)# hasta aqui solo hace este primer grafico

fig.append_trace(
     go.Scatter(
        x=df.index.values,
        y=df['MEATS'],
        name='Carnes'
    ),row=2 , col=1
)

fig.append_trace(
     go.Scatter(
        x=df.index.values,
        y=df['SEAFOOD'],
        name='Productos de mar'
    ),row=3, col=1
)


fig.update_layout(
    height=600,
    width=700,
    title_text="Ventas de 3 tipos de productos"
)

fig.show()
```

```python
import pandas as  pd 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('../data/madrid_metro/estaciones.csv')
df
```

```python
# ojo este dataset nos da la localizacion de las entradas de las estaciones, por lo que tenemos que agrupar
# y cogemos el primer valor de lat y lng por no hacer medias
```

```python
estaciones = df.groupby('DENOMINACION').agg({
    'latitude': 'first', # cogemos la primera 
    'longitude': 'first',
    'ENCUESTADOMICILIARIA': 'sum',
    'GRADOACCESIBILIDAD': 'first' # es una categorica
})

estaciones = estaciones.reset_index()
estaciones
```

```python
import plotly.express as px
```

```python
fig = px.scatter_mapbox(
    estaciones,
    lat='latitude',
    lon='longitude',
    # style
    mapbox_style='carto-positron',
    zoom=10,
    # en vez de mostrar lat y lng mostramso la estacion, hay que resetar el indice
    hover_name='DENOMINACION',
    # definimos la intensidad
    color='ENCUESTADOMICILIARIA',
    color_continuous_scale='matter',

)

#aumentamos el tamaño de los puntos
fig.update_traces(
    marker={
        'size': 12 # no cambia el tamaño con el zoom OJO
    }

)
fig.update_layout( margin={'r':0, 't':0, 'l':0, 'b':0} ) # quitamos el marco blanco
fig.show()

# Min 21:04
```

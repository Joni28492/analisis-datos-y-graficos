adicion de columnas

```python
import random

def calcularGanancias(retweets):
    ganancia = retweets * random.randint(3,5)
    return ganancia

#nueva columnas, le aplicamos la funcion a la columna
df["ganancias"] = df["retweets"].apply(calcularGanancias)


#calcular a partir de varias columnas, Usamos toda la fila
def popularidad(fila):
    resultado = fila["followees"]/fila["followers"]
    return resultado

#es importante el axis=1 para que identifique que lo hace por al fila

df["popularidad"] = df.apply(popularidad, axis=1)

df.head()
```

agrupar los damos

```python
# agrupamos a partir de la columna pais
df.groupby("country").mean() 
# dejara una fila por country
# mean() nos hace el promedio
# a las de texto no les puede sacar el promedio y las quita
```

```python
# aplicando metodo en funcion de columna
grouped = df.groupby("country").agg({
    "followers": "sum",
    "mentions": "mean",
    "retweets": "max"
})

grouped[grouped["followers"] > 5000]
```

```python
grouped.to_csv("salida.csv")
```

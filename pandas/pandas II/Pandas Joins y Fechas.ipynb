```python
import pandas as pd
```

```python
# Carga de archivos
df1 = pd.read_csv("dataset_2a.csv")
df2 = pd.read_csv("dataset_2b.csv")
df3 = pd.read_csv("dataset_2c.csv")
```

el df 1 tiene una col usuario, y el df 2 tiene usuario y ciudad, el df 3 tiene ciudad, es el join de bases de datos, ver los dibujos de los circulos

```python
df_con_usuarios = df1.merge(df2,on="user", how="left") # left, inner etc..
df_con_usuarios
```

```python
df_completo = df_con_usuarios.merge(df3, on="ciudad", how="left")
df_completo
```

las columnas de tipo fecha tienen un formato de tipo string, y no podremos ordenarlo por fechas
referencia formatos fecha
https://docs.python.org/es/3/library/datetime.html#strftime-and-strptime-format-codes


```python
#creamos nueva columna fecha
df_completo["fecha"] = pd.to_datetime( df_completo["fecha_publicado"], format="%d/%m/%Y" )
df_completo
```

ahora ya podemos ordenar por fecha, y transformar la fecha al formato que yo quiera

```python
# stftime string from time
df_completo["dia_semana"] = df_completo["fecha"].apply(lambda x: x.strftime("%A/%d/%B") )
df_completo
```

```python
df_completo["mes"] = df_completo["fecha"].apply(lambda x: x.strftime("%m/%Y") )
df_completo
```

```python
df_completo["mes"] = df_completo["fecha_publicado"].apply(lambda x: x[3:] )
df_completo
```

# Parte 3
## Agrupaciones multiples, pivotar y transponer

```python
## agg es para aplicar una funcion de agregacion
## aqui obtenemos los tweets que hubo por mes
df_completo.groupby('mes').agg({
    'id':'count'
})
```

```python
# hacemos agrupoacion multiple con ciudades para tener subgrupos
agrupado = df_completo.groupby(['mes', 'ciudad']).agg({
    'id':'count'
})#

agrupado
```

 en este caso tiene 2 indices, mes y ciudad, no son columnas

```python
desagrupado = agrupado.reset_index(1) # usamos en indice de nivel 1 para resetar ciudad, 0 seria para mes
desagrupado
```

```python
desagrupado['ciudad']
```

## pivotar
convertir cada valor unico de una columna en nuevas columnas

```python
pivote = desagrupado.pivot(columns='ciudad')
pivote = pivote.droplevel(0, axis='columns') # antes nos salia con el indice id y no nos dejaba acceder a la columna
#asi que lo dropeamos
pivote
```

```python
pivote[['GUAYAQUIL']]
```

Rellenamos los valores nulos o NaN como 0


```python
pivote = pivote.fillna(0)
pivote
```

## Transponer
como en matrices, convertir las columnas en filas y las filas en coumnas


```python
pivote.transpose()
```

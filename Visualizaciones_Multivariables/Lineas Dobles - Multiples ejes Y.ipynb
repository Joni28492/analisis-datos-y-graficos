```python
import pandas as pd
```

```python
df = pd.read_csv('../data/paises_vida.csv')
df = df.dropna()
df_asia=df[df['continent'] == 'Asia']
df_asia_agrupado = df_asia.groupby('year').agg({
    'lifeExp':'mean',
    'gdpPercap':'mean'
})

df_asia_agrupado
```

```python
df_asia_agrupado.plot.line(
    y=['lifeExp', 'gdpPercap'],
    grid=True
) # si nos sale una linea recta es 
# por el problema de rangos muy diferenciados (normalizacion radarchart pero alteramos demasiado)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax1 = plt.subplots()

ax1.plot(
    df_asia_agrupado.index.values, # eje x
    df_asia_agrupado['gdpPercap'],# eje y
    color='red',
    marker='s',
) # hasta aqui tenemos la primera variable

ax1.set_ylabel('GDP Per Capita')
ax1.set_xlabel('Año')
# pintamos numeros del  eje
ax1.tick_params( axis='y', labelcolor='red' )

ax2 = ax1.twinx()# eje gemelo

ax2.plot(
    df_asia_agrupado.index.values, # eje x
    df_asia_agrupado['lifeExp'],# eje y
    color='blue',
    marker='o'
)

ax2.set_ylabel('Expectativa de vida')
ax2.tick_params( axis='y', labelcolor='blue' )


ax1.legend(
    title='',
    labels=['GDP Per Capita'],
    labelcolor=['blue']
    
)

ax2.legend(
    bbox_to_anchor=(0.41,0.93),
    title='',
    labels=['Expectativa de vida'],
    labelcolor=['blue']
    
)

```

en este grafico no hay un patron concreto, solo el incremento pero deberian ser mas parecidas las lineas

# Mapa de calor - HeatMap

### hay diferentes tipos en funcion de las variables
en este caso sera el mas similar al scatterplot

- se utiliza cuando tenemos un scatterplot con demasiados puntos
tambien existe el hexabin con hexagonos


```python
import pandas as pd
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
df.plot.scatter(
    x='flipper_length_mm',
    y='culmen_length_mm',
)
```

```python
import seaborn as sns
```

```python
sns.jointplot(
    data = df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    kind='hist',
    joint_kws={ # configuraciones del grafico central
        'bins': 20
    },
    marginal_kws={ # para los histogramas externos
        'bins':20,
        'fill':False,
    },
    cbar=True, # leyenda degradado

) # recordar calcular coeficiente de correlacion 
```

```python
sns.jointplot(
    data = df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    kind='hex', # para hacer con hexagonos
    joint_kws={ # configuraciones del grafico central
        'bins': 20
    },
    marginal_kws={ # para los histogramas externos
        'bins':20,
        'fill':False,
    },
    #cbar=True, # nos falla el hex
    gridsize=30 # para el tamaÃ±o de los hexagonos
) # recordar calcular coeficiente de correlacion 
```

```python
df.plot.hexbin(
    x='flipper_length_mm',
    y='culmen_length_mm',
    gridsize=30,
)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
import plotly.express as px
```

```python
# para darles color por categoria hay que convertir esas categorias a numeros
df['species_id'] = df['species'].astype('category').cat.codes
df
```

```python
fig = px.parallel_coordinates(
    df,# para darles color por categoria hay que convertir esas categorias a numeros
    color='species_id', # colores por categoria
    dimensions=['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g'],
    color_continuous_scale=px.colors.diverging.Tealrose

)



fig.show()
```

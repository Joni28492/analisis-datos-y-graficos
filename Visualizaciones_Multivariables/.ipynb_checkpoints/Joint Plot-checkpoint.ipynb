# Joint Plot

grafico compuesto - una visualicacion con 2 visualizaciones

variables numericas

```python
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    
)
```

```python
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    hue='species'
)
```

```python
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    kind='reg' # aÃ±ade las lineas de regresion y la de densidad
)
```

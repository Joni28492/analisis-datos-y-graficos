```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tienda_ventas.csv')
df = df.dropna()
df['periodo'] = df['date'].apply(lambda x: x[:7])
df
```

Ids de las tiendas que vamos a utilizar
3, 44, 45, 46, 47, 48, 49, 50

```python
# filtrado
df2 = df[ df['store_nbr'].isin([3, 44, 45, 46, 47, 48, 49, 50]) ].groupby(['periodo', 'store_nbr']).agg({
    'sales':'sum'
})
# con unstack hacemos el pivote, bajamos el nivel y transponemos
matriz = df2.unstack().droplevel(0, axis=1).transpose()
matriz 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(
    figsize=(15,8)
)

sns.heatmap(
    matriz,
    cmap='Blues',
    square=True,
    linewidth=0.1,
    #annot=True # para poner los valores en los cuadraditos, pero queda feo
)
```

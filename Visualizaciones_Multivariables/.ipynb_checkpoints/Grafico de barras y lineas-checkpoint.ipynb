# Grafico de barras y de lineas

## lineas con un segundo eje Y

para ver magnitudes y tendencias

no usar demasiadas lineas

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/paises_vida.csv', index_col='index')
df = df.dropna()
df = df[ df['continent'] == 'Africa' ]
df
```

```python
df_agrupado = df.groupby('year').agg({
    'lifeExp':'mean',
    'pop':'sum',
    'gdpPercap': 'mean'
    
})

df_agrupado = df_agrupado.reset_index() # es importante que empiecen en 0
df_agrupado
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax1 = plt.subplots(
    figsize=(7,3),
)

ax1.grid(visible=True)

sns.barplot(

    data=df_agrupado,
    x='year',
    y='lifeExp',
    color='grey',
    ax=ax1, #hacemos que lo dibuje en ese subplot
    alpha= 0.8,
)

ax2 = ax1.twinx()

# grafico de lineas
sns.lineplot(
    data=df_agrupado['gdpPercap'],
    ax=ax2, 
    color='brown',
    linewidth=2,
    marker='D'
)

ax2.grid(visible=False)
ax2.tick_params(axis='y', labelcolor='brown')
ax2.set_ylabel('GDP Per Capita', color='brown')
```

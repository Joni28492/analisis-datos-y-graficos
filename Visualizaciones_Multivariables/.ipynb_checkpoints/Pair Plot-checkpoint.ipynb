# Pair Plot

es un grafico de todos contra todos,
diversos scaterplot en una cuadricula de nxn
n = numero de varaibles numericas 

en cada fila coge el pivote, y hace la comparacion contra todos

la diagonal es como es contra si mismo, se hace un histograma o kde

para comprobar correlaciones hacerlo el primero

usar pocas columnas

```python
import pandas as pd 
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
sns.pairplot(
    data=df,
)
```

```python
sns.pairplot(
    data=df,
    hue='species'
)
```

```python
sns.pairplot(
    data=df,
    hue='species',
    kind='kde'
)
```

```python
sns.pairplot(
    data=df,
    hue='species',
    kind='hist' # heatmap
)
```

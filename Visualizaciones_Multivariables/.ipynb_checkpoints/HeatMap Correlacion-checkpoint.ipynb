# HeatMap Correlacion
## es una matriz

ambos ejes representan columnas, y dentro de cada casilla tenemos el coeficiente de correlacion
es buen acompañante del pairplot para no calcular estos coeficientes

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
matriz = df[ ['culmen_length_mm', 'culmen_depth_mm','flipper_length_mm','body_mass_g'] ].corr() # coefficiente correlacion lineal
matriz # es una matriz simetrica ojo
```

```python
import seaborn as sns
```

```python
sns.heatmap(
    matriz,
    cmap='Blues',
    square=True,
    annot=True,
)
```

```python
# escalas de colores divergentes
# hacemos que los positivos sean azules y los negativos rojo
sns.heatmap(
    matriz,
    cmap='vlag_r', # la paleta es vlag, pero la revertimos con _r
    square=True,
    annot=True,
)
```

```python
sns.heatmap(
    matriz,
    cmap='coolwarm_r', 
    square=True,
    annot=True,
)
```

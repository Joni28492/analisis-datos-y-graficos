```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
nodos = pd.read_csv('../data/grafo_interaccion/nodes.csv')
nodos
```

```python
ejes = pd.read_csv('../data/grafo_interaccion/edges.csv') 
ejes
```

```python
G = nx.from_pandas_edgelist(

    ejes,
    source='src',
    target='dst',
    edge_attr='weight',
    
)

G # de primeras no te lo pinta
```

```python
plt.rcParams["figure.figsize"] = (15,8) # cambiamnos el tamaÃ±o del canvas


nx.draw(G) # para pintarlo
```

```python
nx.draw_networkx(G)
```

```python
nx.draw_random(G)
```

```python
nx.draw_circular(G)
```

```python
nx.draw_spring(# intenta ponerlo de la mejor manera
    G,
    with_labels=True,
    node_size=300,
    node_color='red',
    alpha=0.8,
    width=1, # ancho de los ejes
    style=':',# estilo de la linea, dash dot etc...
    font_size=10,

)
```

```python
nx.draw_spectral(# nodos mas importantes arriba
    G,
    with_labels=True,
    node_size=300,
    node_color='red',
    alpha=1,
    width=1,
    style=':',
    font_size=10,
   

)
```

```python
nx.draw_spring(# nodos mas importantes arriba
    G,
    with_labels=True,
#     node_size=300,
#     node_color='red',
    alpha=1,
#     width=1,
#     style=':',
    font_size=10,
    
    # tiene que coger las 2 listas para que concuerden en el orden
    nodelist=list(nodos['id']),
    node_size=list(nodos['pagerank'] * 200),
    
    # color por variable categorica
    node_color=list(nodos['community']),
    cmap='tab10',
    
    #ancho de los ejees
    width=list(ejes['weight'] / 7),
    
    
)
```

# Series de Tiempo
es un grafico de lineas, pero se analizan cosas diferentes
- eje x siempre temporal
- eje y orden de 100 1000 etc...
- puede tener tambien una linea de tendencia para realizar predicciones

- revisar estacionalidad, tendencias, perturbaciones etc..


```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('../data/tienda_ventas.csv')
df
```

```python
por_dias = df[['date', 'sales']].groupby('date').sum()
por_dias # ya estan ordenadas
```

```python
por_dias.plot.line(
    y='sales',
    figsize=(30,10),
    
)

plt.title("Ventas por Dia")
```

```python
fig = px.line(

    por_dias,
    x=por_dias.index.values,
    y='sales',
    title="ventas por dias"
)

fig.show()
```

```python
df2 = pd.read_csv('../data/marcas_mensajes_twitter.csv')
df2
```

```python
# tenemos que digerir los datos de fecha, con ayuda de la doc
df2["fecha_parseada"] = pd.to_datetime(df2['created_at'], format='%a %b %d %H:%M:%S %z %Y')
df2
```

```python
df2 = df2.sort_values('fecha_parseada')
df2
```

```python
df2['periodo'] = df2['fecha_parseada'].apply(lambda x: x.strftime('%Y-%m-%d'))
df2
```

```python
df2_agrupado = df2[['periodo', 'idx']].groupby('periodo').count()
df2_agrupado = df2_agrupado.reset_index()
df2_agrupado
```

```python
fig2 = px.line(
    df2_agrupado,
    x='periodo',
    y='idx',
    title='Mensajes por dia'
    

)

fig2.show()
```

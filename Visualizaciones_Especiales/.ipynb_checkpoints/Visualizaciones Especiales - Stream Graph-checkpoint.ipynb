# Stream Graph

es similar al de areas apiladas
el eje y no es un 0
distingue proporcionades entre grupos, no maginutdes.

se recomienda suavizar la linea
eje y = varaible numerica

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/covid_data.csv')
df
```

```python
agrupado = df[['date', 'continent', 'new_cases']].groupby(['date','continent']).sum() # escogemos col y agrupamos
# tenemos que pivotar convertimos los continents en col
agrupado  = agrupado.unstack(1)
# los NaN los configuramos como  0
agrupado = agrupado.fillna(0)
# eliminamos el primer nivel
agrupado = agrupado.droplevel(0, axis=1)
agrupado
```

```python
# construimos una lista de listas
values = [
    agrupado['Africa'],
    agrupado['Asia'],
    agrupado['Europe'],
    agrupado['North America'],
    agrupado['Oceania'],
    agrupado['South America'],
]
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,10)

ax.stackplot(# grafico de areas apiladas
    agrupado.index.values, #eje x
    values, # eje y
    
    baseline='wiggle', # el ejey deja de ser fijo 
    labels=agrupado.columns.values,
    
)


ax.legend( loc="upper left" )
ax.set_title("Casos de Covid por Continente")
ax.set_xlabel('Dia')
ax.set_ylabel('Numero de casos')

# dibijar linea en el 0
ax.axhline(0,color='black', ls='--', linewidth=0.5 )



# formatea eje x con formato fechas
fig.autofmt_xdate()

plt.xticks(np.arange(0,len(agrupado.index.values), 30))



plt.show()
```

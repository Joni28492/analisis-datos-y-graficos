```python
from matplotlib_venn import venn2, venn2_unweighted
from matplotlib_venn import venn3, venn3_unweighted
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../data/animales_caracteristicas.csv')
df
```

```python
respira = set(df[ df["respira"] == 1 ]['animal'])
respira
```

```python
piernas = set(df[ df["piernas"] == 1 ]['animal'])
piernas
```

```python
nada = set(df[ df["nada"] == 1 ]['animal'])
nada
```

```python
venn3(
    [
        respira,
        piernas,
        nada,
    ],
    set_labels=('Respira Aire', 'Tiene Piernas', 'Puede Nadar')
)

plt.title("animales")
```

```python
venn3_unweighted( # mantiene el mismo tamao en todos
    [
        respira,
        piernas,
        nada,
    ],
    set_labels=('Respira Aire', 'Tiene Piernas', 'Puede Nadar')
)

plt.title("animales")
```

## Grafico de barras apiladas


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/datasets_top_4.csv', index_col='dataset')
df
```

```python
df1 = df.iloc[:int(len(df)/2)]
df2 = df.iloc[int(len(df)/2):]
```

```python
fig,(ax1,ax2) = plt.subplots(1,2)
fig.set_size_inches(7,4)


plt.suptitle("Como se distribuyen las categorias\ndentro de los datasets?")



### Primer Subplot

df1.plot.barh(
    stacked=True,
    ax=ax1,
    color={# admite lista tambien
        'cat1': 'darkblue',
        'cat2': 'blue',
        'cat3': 'lightblue',
        'otros':'lightgrey'
    },
    width=0.28
    
)



### Segundo Subplot

df2.plot.barh(
    stacked=True,
    ax=ax2,
    color={
        'cat1': 'darkblue',
        'cat2': 'blue',
        'cat3': 'lightblue',
        'otros':'lightgrey'
    },
    width=0.28
    
)


# si no van debajo del plot no funcionan correctamente
ax1.set_ylabel('Dataset')
ax1.set_xlabel('Porcentaje')
ax1.legend().remove()

ax2.set_ylabel('')
ax2.set_xlabel('Porcentaje')
ax2.yaxis.tick_right() # ponemos los ticks hacia la dcha

ax2.legend(
    ['Categoria 1','Categoria 2','Categoria 3','Otros'],
    ncol=4,
    loc='upper left',
    bbox_to_anchor=(-1.3,-0.15)
)


ax1.text(
    75, -5, # son puntos referentes al eje X del ax1
    'Algunos datasets del analisis han sido ignorados',
    fontsize=6,
    bbox={
        "facecolor":'orange', # color de fondo
        "alpha":0.5
    }
    
)




```

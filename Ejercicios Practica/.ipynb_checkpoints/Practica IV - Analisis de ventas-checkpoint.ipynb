## Analisis de ventas

¿Cuales son las tiendas que han tenido un mayor numero de ventas historicamente y como ha sido su evolucion a lo largo del tiempo?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/tienda_ventas.csv')
df
```

Cuantas tiendas tenemos

```python
len(df['store_nbr'].unique())
```

```python
 len(df['date'].unique()) # el dataset ya esta ordenado
```

```python
df['mes'] = df['date'].apply(lambda x: x[:7]) #obtenemos el año y mes para reducir el espacio de analisis
df
```

```python
len(df['mes'].unique())
```

Buscamos las tiendas que mas ventas de productos tienen para tener un top N

```python
agrupado_tiendas = df.groupby("store_nbr").agg({
    'sales':'sum'
})

agrupado_tiendas
```

```python
agrupado_tiendas.describe()
```

```python
agrupado_tiendas.plot.box()
```

```python
agrupado_tiendas.plot.hist(bins=20)
```

```python
agrupado_tiendas.sort_values('sales', ascending=False) # ordenamos para verel top
```

```python
top_5_tiendas = [44, 45, 47, 3, 46]
```

```python
df_top_5 = df[df['store_nbr'].isin(top_5_tiendas)]  #filtramos el df original
df_top_5
```

Haremos un grafico de lineas de las 5 tiendas

```python
top_5_por_mes = df_top_5.groupby(['store_nbr', 'mes']).agg({
    'sales':'sum'
})

top_5_por_mes = top_5_por_mes.reset_index()
top_5_por_mes
```

```python
ax = sns.lineplot(
    data=top_5_por_mes,
    x='mes',
    y='sales',
    hue= 'store_nbr',
    style= 'store_nbr', # le da un estilo diferente a cada linea dotted dashed etc...
    markers=True,
#     palette='Set1'
)

ax.tick_params(
    axis='x',
    labelrotation=90,
    labelsize=8
)

ax.tick_params(
    axis='y',
    labelsize=8
)

ax.grid(
    visible=True,
    axis='x',
    alpha=0.2
)
ax.grid(
    visible=True,
    axis='y',
    alpha=0.2
)

ax.legend(
    title='Top 5 Tiendas',
    fontsize=8
)

ax.set_ylabel('Ventas')
ax.set_xlabel('Mes')

#cambiar el tamaño de la figura localmente
fig = ax.get_figure()
fig.set_size_inches(11,3)
```

Vamos a dibujar el grafico linea por linea, graficacioniterativa

```python
estilos_por_tienda = {# color, alpha, marcadores, estilo de la linea, ancho linea
    44:['grey',0.4, '+', '--',1],
    45:['grey',0.4, 's', '--',1],
    47:['grey',0.4, 'x', '--',1],
    3:['red',1.0, 'o', '-',1.5],
    46:['grey',0.4, 'D', '--',1]
}
```

```python
# creamos un canvas vacio
fig , ax2 = plt.subplots()
```

```python
# creamos un canvas vacio
fig , ax2 = plt.subplots()

# epezamos a dibujar sobre el 
for tienda in top_5_tiendas:
    estilos = estilos_por_tienda[tienda]
    df_tmp = top_5_por_mes[top_5_por_mes['store_nbr'] == tienda]# df temporal
    
    ax2.plot(
        df_tmp['mes'],
        df_tmp['sales'],
        # personalizacion
        label=tienda,
        markersize=4,
        # personalizacion diccionario
        color =estilos[0],
        alpha =estilos[1],
        marker =estilos[2],
        linestyle=estilos[3],
        linewidth=estilos[4]
    )

    
# personalizacion grafico anterior
ax2.tick_params(
    axis='x',
    labelrotation=90,
    labelsize=8
)

ax2.tick_params(
    axis='y',
    labelsize=8
)

ax2.grid(
    visible=True,
    axis='x',
    alpha=0.2
)
ax2.grid(
    visible=True,
    axis='y',
    alpha=0.2
)

ax2.legend(
    title='Top 5 Tiendas',
    fontsize=8
)

ax2.set_ylabel('Ventas')
ax2.set_xlabel('Mes')


fig.set_size_inches(11,3)
```
